## 4.1 Циклы с заданным числом повторений



```python
print("Проверка формулы x² + x + 17 для x от 0 до 15")

all_prime = True

for x in range(0, 16):
    n = x*x + x + 17
    
    # Проверка на простоту
    is_prime = True
    if n < 2:
        is_prime = False
    else:
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                is_prime = False
                break
    
    if not is_prime:
        all_prime = False
    
    print(f"x = {x:2}: {n} {'(простое)' if is_prime else '(НЕ простое)'}")


if all_prime:
    print("Все числа простые!")
else:
    print("Ошика! Есть составные числа")
```

## 4.2 Суммы и произведения числовых последовательностей



```python
a = float(input("Введите действительное число a: "))
n = int(input("Введите натуральное число n: "))

if n <= 0:
    print("Ошибка, n должно быть натуральным")
else:
    terms = []
    for k in range(n + 1):
       terms.append(a - k * n)
    P = 1
    for t in terms: 
        P *= t
    
    print(f"\na = {a}, n = {n}")
    print(f"P = {' * '.join([f'({t:.2f})' for t in terms])}")
    print(f"P = {P:.4f}")
```
