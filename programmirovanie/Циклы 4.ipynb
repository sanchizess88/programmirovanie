## 4.1 Циклы с заданным числом повторений



```python
import math

epsilon = float(input("Введите число epsilon (например, 0.01): "))

n = 1           

total_sum = 0.0 

# Запускаем бесконечный цикл
while True:
    numerator = (2 ** n) * math.factorial(n)
    
    # Вычисляем знаменатель по формуле (n в степени n).
    denominator = n ** n
    
    # Считаем текущий член ряда a_n, разделив числитель на знаменатель.
    a_n = numerator / denominator
    
    # Проверяем главное условие: если модуль (значение без знака) a_n >= epsilon:
    if abs(a_n) >= epsilon:
        total_sum += a_n
        n += 1
    else:
        # Если значение a_n стало слишком маленьким (меньше epsilon), 
        # значит, дальше считать нет смысла — выходим из цикла.
        break

# Выводим результат на экран.
print(f"Итоговая сумма: {total_sum}")

# Выводим n - 1, потому что последнее n, на котором цикл прервался, мы уже не прибавляли.
print(f"Количество сложенных элементов: {n - 1}")
```

## 4.2 Суммы и произведения числовых последовательностей



```python
import math

a = float(input("Введите действительное число a: "))
n = int(input("Введите натуральное число n: "))

if n <= 0:
    print("Ошибка, n должно быть натуральным")
else:
    # Генерируем элементы сразу внутри списка
    terms = [a - k * n for k in range(n + 1)]
    
    # math.prod для быстрого умножения списка
    P = math.prod(terms)
    
    print(f"\na = {a}, n = {n}")
    print(f"P = {' * '.join([f'({t:.2f})' for t in terms])}")
    print(f"P = {P:.4f}")
```

# 4.3 Итерационные циклы

```python
import math

epsilon = float(input("Введите число epsilon (например, 0.01): "))

n = 1           # Начинаем с первого члена ряда
total_sum = 0.0 # Переменная для хранения итоговой суммы

while True:
    # Считаем числитель: 2 в степени n умножить на факториал n
    numerator = (2 ** n) * math.factorial(n)
    # Считаем знаменатель: n в степени n
    denominator = n ** n
    
    # Вычисляем текущий член ряда a_n
    a_n = numerator / denominator
    
    # Проверяем: если модуль a_n больше или равен epsilon, то прибавляем к сумме
    if abs(a_n) >= epsilon:
        total_sum += a_n
        n += 1 # Переходим к следующему члену ряда
    else:
        # Если a_n стало меньше epsilon, останавливаем цикл
        break

print(f"Итоговая сумма: {total_sum}")
print(f"Количество сложенных элементов: {n - 1}")
```

# 4.4 Табулирование функций

```python
import math

# Ввод исходных данных
a = float(input("Введите начало отрезка a: "))
b = float(input("Введите конец отрезка b: "))
h = float(input("Введите шаг h: "))

print("    x     |     F(x)    ")

x = a

# Цикл работает, пока x не дойдет до конца отрезка b.
# Добавляем маленькое число 1e-9 (0.000000001), чтобы избежать 
# ошибок округления дробей в Python и точно включить правую границу b.
while x <= b + 1e-9:
    # Защита от деления на ноль
    if x == 0:
        print(f"{x:9.2f} | Ошибка (деление на 0)")
    else:
        # Считаем значение функции
        f_x = math.sin(1 / x) + 2
        # Выводим результат в таблицу (округляем до 4 знаков после запятой)
        print(f"{x:9.2f} | {f_x:10.4f}")
        
    x += h # Делаем шаг: увеличиваем x на значение h
```
