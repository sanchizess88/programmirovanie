Оператор выбора

```python
import math

# Спрашиваем у пользователя
print("Какой элемент треугольника известен?")
print("1 - катет (a)")
print("2 - гипотенуза (b)")
print("3 - высота (h)")
print("4 - площадь (S)")

choice = int(input("Введите номер фигуры (1-4): "))
value = float(input("Введите значение: "))

# Проверка
if choice < 1 or choice > 4:
    print("Ошибка! Неправильный номер")
elif value <= 0:
    print("Ошибка! Значение должно быть положительным")
else:
    # Вычисления
    if choice == 1:  # катет
        a = value
        b = a * math.sqrt(2)
        h = a / math.sqrt(2)
        S = a * a / 2
        
    elif choice == 2:  # гипотенуза
        b = value
        a = b / math.sqrt(2)
        h = b / 2
        S = a * a / 2
        
    elif choice == 3:  # высота
        h = value
        a = h * math.sqrt(2)
        b = a * math.sqrt(2)
        S = a * a / 2
        
    elif choice == 4:  # площадь
        S = value
        a = math.sqrt(2 * S)
        b = a * math.sqrt(2)
        h = a / math.sqrt(2)
    
    # Вывод результатов
    print("\nРезультаты")
    print(f"Катет (a) = {round(a, 3)}")
    print(f"Гипотенуза (b) = {round(b, 3)}")
    print(f"Высота (h) = {round(h, 3)}")
    print(f"Площадь (S) = {round(S, 3)}")
```
